```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/sagark721/Datasets/main/Train.csv")
```

```python
df.head()
```

```python
#Lets Convert Outlet_Establishment_Year into object type:

df['Outlet_Establishment_Year']=df['Outlet_Establishment_Year'].astype(object)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
cat_df= df.select_dtypes(include=object)
```

```python
num_df=df.select_dtypes(exclude=object)
```

```python
cat_df.columns
```

```python
plt.figure(figsize=(10,7),dpi=101)

for col in cat_df.columns:
    sns.displot(cat_df[col],)
    plt.tight_layout()
    plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(10,7),dpi=101)
for col in num_df.columns:
    sns.displot(num_df[col],kde=True)
plt.tight_layout()
plt.xticks(rotation=90);
```
